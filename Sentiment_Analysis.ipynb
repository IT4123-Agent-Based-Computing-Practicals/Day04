```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tokenize import word_tokenize

nltk.download("punkt")
nltk.download("stopwords")
nltk.download("wordnet")
nltk.download("omw-1.4")
```

```python
positive_words = ["love", "fantastic", "great", "amazing", "happy", "good"]
negative_Words = ["worst", "disappointed", "terrible", "bad", "awful", "poor"]

lemmatizer = WordNetLemmatizer()

def preprocess_text(text):
    text = text.lower()
    
    words = word_tokenize(text)

    stop_words = set(stopwords.words("english"))
    filtered_words = [word for word in words if word not in stop_words]

    lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]

    return lemmatized_words

def analyze_sentiment(text):
    words = preprocess_text(text)

    positive_count = sum(1 for word in words if word in positive_words)
    negative_count = sum(1 for word in words if word in negative_Words)

    if positive_count > negative_count:
        return "possitive"
    elif negative_count < positive_count:
        return "negative"
    else:
        return "neutral"
```

```python
data = [
    {"text":"I love this product!"},
    {"text":"This is the worst service I've ever experienced!"},
    {"text":"This movie was okay, not great but not terrible!"},
    {"text":"What a fantastic day!"},
    {"text":"I am very disappointed with the quality!"}
]
```

```python
for entry in data:
    text = entry["text"]
    sentiment = analyze_sentiment(text)
    print(f"Text: {text} => Sentiment: {sentiment}")
```
